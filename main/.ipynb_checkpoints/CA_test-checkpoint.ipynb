```python
# module 0: merge dam and neg data

# Import necessary libraries
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from scipy.spatial import cKDTree
import matplotlib.pyplot as plt

# Merge CA_df_dam.csv and CA_df_neg.csv
def merge_dam_and_neg_locations(df_dams, df_neg):
    neg_coords = df_neg[['Longitude', 'Latitude']].values
    neg_tree = cKDTree(neg_coords)

    dam_coords = df_dams[['longitude', 'latitude']].values
    distances, indices = neg_tree.query(dam_coords)

    df_dams['neg_Longitude'] = df_neg.iloc[indices]['Longitude'].values
    df_dams['neg_Latitude'] = df_neg.iloc[indices]['Latitude'].values
    return df_dams


# create neg damid and inherit other
def create_negative_locations(df_dams):
    neg_locations = df_dams.copy()
    neg_locations['longitude'] = df_dams['neg_Longitude']
    neg_locations['latitude'] = df_dams['neg_Latitude']
    neg_locations['damid'] = df_dams['damid'].apply(lambda x: f"{x}n")
    neg_locations['geometry'] = gpd.points_from_xy(neg_locations['longitude'], neg_locations['latitude'])
    return neg_locations

# Merge positive and negative location
def merge_positive_and_negative(df_dams, neg_locations):
    combined = pd.concat([df_dams, neg_locations]).sort_index(kind='merge')
    combined = combined.reset_index(drop=True)
    return combined

# Calculate and plot the distribution of negative and positive locations
def plot_location_distribution(df_dams, neg_locations):
    gdf_pos = gpd.GeoDataFrame(df_dams, geometry=gpd.points_from_xy(df_dams.longitude, df_dams.latitude))
    gdf_neg = gpd.GeoDataFrame(neg_locations, geometry=gpd.points_from_xy(neg_locations.longitude, neg_locations.latitude))

    # Plot the locations
    fig, ax = plt.subplots(figsize=(10, 8))
    gdf_pos.plot(ax=ax, color='blue', alpha=0.5, markersize=5, label='Positive Locations')
    gdf_neg.plot(ax=ax, color='red', alpha=0.7, markersize=5, label='Negative Locations')
    plt.legend()
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Distribution of Positive and Negative Locations')
    plt.grid(True)
    plt.show()
```

```python
# Main
dam_data_path = 'data/CA_df_dam.csv'
neg_data_path = 'data/CA_df_neg.csv'
time_window = timedelta(days=3)
space_buffer = 0.1

# preprocess data
gdf_dams = pd.read_csv(dam_data_path)
gdf_neg = pd.read_csv(neg_data_path)

# Merge location
gdf_dams = merge_dam_and_neg_locations(gdf_dams, gdf_neg)
gdf_dams.head()

# Create negative location dataframe
neg_locations = create_negative_locations(gdf_dams)

# Combine locations
combined_df = merge_positive_and_negative(gdf_dams, neg_locations)
print(combined_df)

# Plot Location Distribution
plot_location_distribution(gdf_dams, neg_locations)
```

```python
merged_file = 'data/CA_merged.csv'
combined_df.to_csv(merged_file, index=False)

print(f"Results saved to {merged_file}")
```

```python
# module 2data preprocess
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd

def preprocess_data(dam_data_path):
    df_dams = pd.read_csv(dam_data_path)
    df_dams['date'] = pd.to_datetime(df_dams['date'])

    geometry = [Point(xy) for xy in zip(df_dams['longitude'], df_dams['latitude'])]
    gdf_dams = gpd.GeoDataFrame(df_dams, geometry=geometry)
    
    return gdf_dams
```

```python
# module 3define search function
def query_satellite_images(m2m, start_date, end_date, bbox, max_cloud_cover=10, dataset_name="landsat_etm_c2_l2", max_results=100):
    params = {
        "datasetName": dataset_name,
        "startDate": start_date.strftime('%Y-%m-%d'),
        "endDate": end_date.strftime('%Y-%m-%d'),
        "boundingBox": (bbox[1], bbox[0], bbox[3], bbox[2]),  # boundingBox (minY, minX, maxY, maxX)
        "maxResults": max_results,
        "maxCC": max_cloud_cover
    }
    
    scenes = m2m.searchScenes(**params)
    print("{} - {} hits - {} returned".format(dataset_name, scenes['totalHits'], scenes['recordsReturned']))
    return scenes
```

```python
# module 4merge dam data
from datetime import timedelta
from shapely.geometry import box

def merge_dam_data(gdf, time_window, space_buffer):
    merged_data = []
    
    while not gdf.empty:
        ref_point = gdf.iloc[0]
        ref_date = ref_point['date']
        
        time_range = (gdf['date'] >= ref_date - time_window) & (gdf['date'] <= ref_date + time_window)
        space_range = gdf.geometry.apply(lambda x: x.within(ref_point.geometry.buffer(space_buffer)))
        
        in_range = gdf[time_range & space_range]
        
        minx, miny, maxx, maxy = in_range.total_bounds
        merged_bbox = box(minx, miny, maxx, maxy)
        merged_start_date = in_range['date'].min()
        merged_end_date = in_range['date'].max()
        
        merged_data.append({
            'bbox': merged_bbox,
            'start_date': merged_start_date,
            'end_date': merged_end_date
        })
        
        gdf = gdf.drop(in_range.index)
    
    return merged_data
```

```python
# module 5search identifiers and add to DataFrame
def query_and_collect_display_ids(gdf_dams, merged_data, m2m):
    all_results = []
    
    for data in merged_data:
        start_date = data['start_date']
        end_date = data['end_date']
        bbox = data['bbox'].bounds

        original_dams = gdf_dams[
            (gdf_dams['date'] >= start_date) & 
            (gdf_dams['date'] <= end_date) & 
            (gdf_dams.geometry.within(box(*bbox)))
        ]
        
        print(f"Querying satellite images from {start_date} to {end_date} within bbox {bbox}")
        
        scenes = query_satellite_images(m2m, start_date, end_date, bbox)
        
        if scenes and 'results' in scenes:
            print(f"Found {len(scenes['results'])} scenes")
            for idx in original_dams.index:
                #  `displayId`
                display_id = scenes['results'][0]['displayId']
                all_results.append({
                    'index': idx,
                    'displayId': display_id
                })
        else:
            print("No scenes found for this query")
    
    results_df = pd.DataFrame(all_results)
    gdf_dams = gdf_dams.join(results_df.set_index('index'), on=gdf_dams.index.name)
    
    return gdf_dams
```

```python
# search displayId and add to DataFrame
gdf_dams_with_display_ids = query_and_collect_display_ids(gdf_dams, merged_data, m2m)
print(gdf_dams_with_display_ids)
```
